```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
countries = pd.read_csv('./countries_val.csv')
countries['Year'] = pd.to_datetime(countries['Year'],format='%Y')
countries['Year'] = pd.DatetimeIndex(countries['Year']).year
countries["Value"] = pd.to_numeric(countries["Value"])
```

```python
format_dict = {'Value': '{0:,.0f}fl oz'}
group = countries.groupby(["Country","Year"])[["Value"]].sum()
#total_sup = pd.DataFrame(total_group)
group.style.format(format_dict)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=group, x='Year', y='Value', hue='Country', style= 'Country')

plt.title("Per Capita Consumption of Coca-cola Per Country")
```

```python
countries.groupby(["Year"])[["Value"]].sum().plot.bar(color = 'teal')
plt.title("Per Capita Consumption of Coca-cola 2009 - 2012")
plt.show()
```

```python
financials = pd.read_csv('./financials.csv')
financials["Value"] = pd.to_numeric(financials["Value"])
```

```python
format_dict = {'Value': '${0:,.0f} mil'}
fin_grp = financials.groupby(["Item","Year"])[["Value"]].sum()
#total_sup = pd.DataFrame(total_group)
fin_grp.style.format(format_dict)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=fin_grp, x='Year', y='Value', hue='Item', style= 'Item')
plt.title("Financial Highlights of Coca-cola 2009 - 2019")
```
